### Task 1 - DataFrame manipulation

Using **states_covid.csv**, we are going to read the data in as a DataFrame to manipulate, subset, and filter in various ways. 

**1.1 Read in states_covid.csv with date as a "date" dtype, and only columns consisting of the hospitalization (4 col), ICU (2 col), and Ventilators (2 col)**

```python
#I am creating a datafram with only the date, dtype, and columns 
#consisting of hospitalization, ICU and ventilators.
#Then I am printing it to the screen with the "head" feature to only show 5 at a time. 
state_covid_df = pd.read_csv('states_covid.csv',usecols=['date', 'state', 'hospitalized', 'hospitalizedCumulative','hospitalizedCurrently', 'hospitalizedIncrease','inIcuCumulative','inIcuCurrently','onVentilatorCumulative', 'onVentilatorCurrently'],parse_dates=['date'],infer_datetime_format=True)
state_covid_df.head()
```

**1.2 List, in order, the 5 states with the highest numbers of people *currently* hospitalized, in the ICU, and on ventilation**  
hint: sort_values, uniqie

```python
#First I organized by current status overall 
state_covid_df.sort_values(by=['onVentilatorCurrently','hospitalizedCurrently', 'inIcuCurrently'], ascending=False)
state_covid_df.head()
```

```python
#Next, I created a new dataframe called state_df.
#I organized by cumulative state means per category
#This created stacked MultiIndex column names. 
#I printed to the screen using .head() the top 5 results. 
state_df = state_covid_df.groupby('state', as_index=False).agg(['mean'])
state_df.head()
```

```python
#To get rid of the stated MultiIndex column names, I joined the mean to the original column name.
#I printed to the screen using .head() the top 5 results. 
state_df.columns = state_df.columns.map(''.join)
state_df.head()
```

```python
#For my final step, I sorted my state_df dataframe by the "currently" columns. 
##I also printed to the screen using .head() the top 5 results. 
#This shows me the top 5 states that have the highest "current" means.
state_df.sort_values(by=['onVentilatorCurrentlymean','hospitalizedCurrentlymean', 'inIcuCurrentlymean'], ascending=False).head()
```

## Task 2 - DataFrame summarizing

Using **Bloom_etal_2018_Reduced_Dataset.csv**, we are going to do more dataframe manipulation and subsetting and summarizing  

**2.1 Read in Bloom_etal_2018_Reduced_Dataset.csv and create two new columns ('genus','species') that consists of the column *taxa* split at the underscore. Print out the head of this new dataframe and the number of unique genera**   

*hint:* pd.str.split(,expand=True)

for example:

| taxa | genus | species 
| :------ | :-- | :--- 
| Alosa_alabamae | Alosa | alabamae

```python
#I am creating a dataframe to easily read the file. 
bloom_df = pd.read_csv('Bloom_etal_2018_Reduced_Dataset.csv') 
#I am creating a new dataframe to split the previous column taxa into two separate columns.
new = bloom_df['taxa'].str.split("_", n=1, expand = True)
#I am splitting the first column into 'genus'
bloom_df['genus']= new[0]
#I am splitting the second column into 'species'
bloom_df['species']= new[1]
#I also printed to the screen using .head() the top 5 results. 
bloom_df.head()
```

```python
#I wanted to see how many different genera there are in this file.
len(bloom_df['genus'].unique())
```

**2.2 Create a new dataframe with the mean *logbodysize* and *trophicposition* of each genera. Sort this data frame by the largest body size. Print the head of this dataframe. Which genera is the smallest and largest? What is the trophic position of the smallest and largest?**

```python
#I am copying what I did above and creating a new dataframe that will have the mean of each genus' logbodysize and trophic_position. 
#I printed to the screen using .head() the top 5 results. 


bloom_mean_df = bloom_df.groupby('genus', as_index=False).agg(['mean'])
bloom_mean_df.head()

#bloom_df = bloom_df.sort_values(by=['genus','logbodysize','trophic_position'])
#print(bloom_df)
```

```python
#Next, I did what I did above and got rid of the stated MultiIndex column names.
#I joined the mean to the original column name.
#I printed to the screen using .head() the top 5 results. 
bloom_mean_df.columns = bloom_mean_df.columns.map(''.join)
bloom_mean_df.head()
```

```python
#Next, I am organizing the top values by log body size. 
bloom_mean_df.sort_values(by=['logbodysizemean'], ascending=False).head()

#We can see that the largest genus is Tenualosa. 
#Its log body size mean is 1.778 and its tropic position is 0.4624.
```
